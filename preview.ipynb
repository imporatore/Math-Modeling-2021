```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from utils import DATA_DIR
```

```python
full_data = pd.read_excel(os.path.join(
    DATA_DIR, "2021MCMProblemC_DataSet.xlsx"),
    parse_dates=["Submission Date"])
image_data = pd.read_excel(os.path.join(
    DATA_DIR, "2021MCM_ProblemC_ Images_by_GlobalID.xlsx"))
image_data = pd.merge(image_data, full_data[['GlobalID', "Lab Status"]], how='left')
```

```python
# pos_mask = (full_data['Lab Status'] == 'Positive ID')
# neg_mask = (full_data['Lab Status'] == 'Negative ID')
# unv_mask = (full_data['Lab Status'] == 'Unverified')
# unp_mask = (full_data['Lab Status'] == 'Unprocessed')

pos_mask, neg_mask, unv_mask, unp_mask = [(
        full_data["Lab Status"] == i) for i in
    ["Positive ID", "Negative ID", "Unverified", "Unprocessed"]]
```

```python
plt.scatter(full_data.loc[pos_mask, 'Longitude'],
            full_data.loc[pos_mask, 'Latitude'],
            c=full_data.loc[pos_mask, 'Detection Date'].apply(
                lambda x: (x - full_data.loc[pos_mask, 'Detection Date'].min()).days))
for x, y, s in zip(full_data.loc[pos_mask, 'Longitude'],
                   full_data.loc[pos_mask, 'Latitude'],
                   full_data.loc[pos_mask, 'Detection Date'].apply(
                       lambda x: x.month)):
    plt.text(x, y, s)
plt.plot(full_data.copy().loc[pos_mask, :].sort_values('Detection Date')['Longitude'],
         full_data.copy().loc[pos_mask, :].sort_values('Detection Date')['Latitude'])
plt.colorbar()
```
